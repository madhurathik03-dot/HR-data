**SUMMARY**

Dataset Overview: The dataset hr currently contains information on 1423 employees across 41 columns after dropping rows with missing values and performing feature engineering. It includes a mix of numerical (e.g., Age, DailyRate, MonthlyIncome) and categorical features (e.g., Department, JobRole, Gender).

Initial Data Loading and Inspection:

The dataset was loaded from /content/HR_Analytics.csv into a pandas DataFrame named hr.
Basic checks like hr.head(), hr.info(), and hr.shape were performed to understand its structure.
Missing Values and Data Cleaning:

Initially, the YearsWithCurrManager column was found to have 57 missing values.
The WorkLifeBalance column was dropped.
All rows containing missing values were subsequently dropped using hr.dropna(inplace=True), resulting in 1423 remaining entries and no more nulls.
Statistical Information:

hr.describe() provided statistical summaries for numerical columns.
Key observations included Age ranging from 18 to 60, and constant values for EmployeeCount and StandardHours.

Also aplied the logic showing which employee id earn maximum and minimum monthly income

Feature Engineering: Several new features were created:

Income_Difference: Calculated as MonthlyIncome - MonthlyRate.
Stable_Income: A binary indicator (1 or 0) showing if MonthlyIncome is greater than or equal to MonthlyRate.
Expected_DailyRate: Calculated as HourlyRate * HOURS_PER_DAY (where HOURS_PER_DAY was set to 8).
Daily_Hourly_Ratio: Calculated as DailyRate / HourlyRate.
Categorical Data Insights:

Departments: The dataset primarily consists of employees from 'Research & Development' (967), 'Sales' (450), and 'Human Resources' (63).
Business Travel: A slight inconsistency was noted with 'TravelRarely' (8 occurrences) possibly being a misspelling of 'Travel_Rarely' (1042 occurrences).
Job Roles: There are 9 unique job roles, with 'Sales Executive' and 'Research Scientist' being among the most frequent.
Education Field: 'Life Sciences', 'Medical', and 'Marketing' are the most common education fields.
Visualizations Performed:

Gender Distribution by Department: A count plot showed the number of male and female employees across different departments.
Marital Status Distribution: A horizontal bar chart visualized the count of employees by their marital status.
Department and Job Role Counts: Bar charts displayed the total number of employees in each department and job role.
Education Field Distribution: A bar plot highlighted the number of employees in various education fields.
Income Difference Distribution: A KDE plot provided insight into the distribution of the Income_Difference.
Proportion of Employees by Job Role: A pie chart illustrated the percentage breakdown of employees across job roles.
Monthly Income per Department: A bar chart (with adjustable bar width) compared the average monthly income across the 'Human Resources', 'Research & Development', and 'Sales' departments.
